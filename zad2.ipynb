```
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras import *
from math import *
 
```

```
def Funkcja_1d(x):
  return (x-2)*(x+1)*(x-4)

def Funkcja_2d(x,y):
  z = sin(sqrt((x*x)+(y*y)))/sqrt((x*x)+(y*y))
  return z
```

```
print("Zbiór x w funkcji 1d")
x1 = np.arange(-15,15,0.5,dtype=float)

print(x1)

print("Zbiór y w funkcji 1d")


y1 = np.array([Funkcja_1d(i) for i in x1])

print(y1)

print("Zbiór x w funkcji 2d")

x2 = np.arange(-2, 2, 0.25, dtype=float)

print(x2)
print("Zbiór y w funkcji 2d")
y2 = np.arange(-2.25, 1.75, 0.25, dtype=float)

print(y2)

print("Zbiór z w funkcji 2d")
z =[]
for i in range(len(x2)):
    z.append(Funkcja_2d(x2[i],y2[i]))
z = np.array(z)
print(z)
```

```
# model1_d = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=(1,))
                               
# ])

model1_d = tf.keras.Sequential([
 keras.layers.Dense(units=1, input_shape=(1,)),
  keras.layers.Dense(units = 5, activation=keras.activations.relu),
  keras.layers.Dense(units = 1)
])


#model1_d.compile(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))#(optimizer='sgd', loss='mean_squared_error')#(loss='mse',optimizer=tf.optimizers.Adam(learning_rate=0.1))
model1_d.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.5),
    loss = keras.losses.mse,
    metrics = ['accuracy']
)
```

```
model1_d.fit(x1, y1, epochs=500,shuffle=True,validation_split=0.01)
```

```
print("Tyle wyszło:")
print(float(model1_d.predict([20.0])))
print("Tyle powinno wyjść:")
print(Funkcja_1d(20))
print("Tyle wyszło:")
print(float(model1_d.predict([4.0])))
print("Tyle powinno wyjść:")
print(Funkcja_1d(4))

print("Tyle wyszło:")
print(float(model1_d.predict([15.0])))
print("Tyle powinno wyjść:")
print(Funkcja_1d(15))
```

Model dla funkcji 1d jest dokładniejszy w przypadku liczb znajdujących się w zbiorze treninigowym,walidacja krzyżowa zwiększa dokładność modelu

```
model_2d = keras.Sequential([
  keras.layers.Dense(units=2, input_shape=(2,)),
  keras.layers.Dense(units = 10, activation=keras.activations.hard_sigmoid),
  # keras.layers.Dense(units = 3, activation=keras.activations.sigmoid),
  keras.layers.Dense(units = 10, activation=keras.activations.relu),
  keras.layers.Dense(units = 1)
])
model_2d.summary()
```

Tworzę model z jedną "ukrytą" warstwą

```

x_y = np.stack([x2,y2], axis=1)
model_2d.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.1),
    loss = 'mean_squared_error'
)
```

```
model_2d.fit(x_y, z, epochs=100,shuffle=True,validation_split=0.001)
```
